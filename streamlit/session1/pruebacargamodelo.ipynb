```python
import torch
from torchvision import models
import torch.nn as nn

# Define y carga el modelo correctamente antes de intentar cargar su estado
def load_model():
    # Crear la instancia del modelo usando la arquitectura ResNet50
    model = models.resnet50(pretrained=False)
    
    # Obtener el número de características del último layer lineal (Fully Connected layer)
    num_ftrs = model.fc.in_features
    
    # Reemplazar la última capa para coincidir con el número de clases del modelo entrenado
    model.fc = nn.Linear(num_ftrs, 15)  # Asegúrate de cambiar el 15 por el número correcto de clases si es necesario

    # Cargar el estado del modelo desde el archivo, asegurándote de especificar la ubicación del mapa para el dispositivo correcto
    model.load_state_dict(torch.load('models/model_1_grande.pth', map_location=torch.device('cpu')))
    
    # Colocar el modelo en modo de evaluación para desactivar capas como Dropout y BatchNorm durante la inferencia
    model.eval()

    return model

# Llama a la función para cargar el modelo
model = load_model()

# Opcional: imprime parte del modelo para confirmar la carga correcta
print(model)
```
